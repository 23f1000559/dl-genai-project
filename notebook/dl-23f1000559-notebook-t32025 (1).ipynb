```python
#tokenisation
import torch
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
```

```python
df = pd.read_csv('/kaggle/input/2025-sep-dl-gen-ai-project/train.csv')
test = pd.read_csv('/kaggle/input/2025-sep-dl-gen-ai-project/test.csv')
```

```python
df.head()
```

```python
from datasets import Dataset
```

```python
test_df = pd.read_csv("/kaggle/input/2025-sep-dl-gen-ai-project/test.csv")
test_dataset = Dataset.from_pandas(test_df)
```

```python
cols = ["anger","fear","joy","sadness","surprise"]
labels = df[cols].astype(float).values
```

```python
dataset = Dataset.from_dict({
    'text': df['text'].tolist(),
    'labels' : labels.tolist()
})
```

```python
dataset
```

```python
dataset = dataset.train_test_split(test_size=0.1, seed=42)
train_dataset = dataset["train"]
val_dataset = dataset["test"]
```

# tokenisation


```python
model_name = "roberta-base" 
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(batch):
    return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=128)


train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset = val_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)
```

```python
train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
val_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
#Model 
num_labels = len(train_dataset[0]["labels"]) 
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=5,
    problem_type="multi_label_classification"
)
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    report_to = "wandb",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=2e-5,
    num_train_epochs=2,
    logging_steps=200,
    gradient_accumulation_steps=1
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
)

trainer.train()
```

```python
preds = trainer.predict(test_dataset)
probs = torch.sigmoid(torch.tensor(preds.predictions)).numpy()

if probs.shape[0] == 5:  
    probs = probs.T

binary_preds = (probs > 0.5).astype(int)
```

```python
# submission
output = pd.DataFrame(binary_preds, columns=["anger", "fear", "joy", "sadness", "surprise"])
output.insert(0, "id", test_df["id"])
output.to_csv("submission.csv", index=False)
```
